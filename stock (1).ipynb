# Apple

```python
import yfinance as yf
```

```python
apple_ticker = yf.Ticker("AAPL")
# get stock info
# get historical market data
apple_df= apple_ticker.history(period="6y")
apple_df["Change"] = (apple_df["Close"] - apple_df["Open"]).map("{:.2f}".format)
apple_df["%Change"] = ((apple_df["Close"] - apple_df["Open"])/apple_df["Open"]*100).map("{:.2f}%".format)

apple_df
```

```python
# find the change of stocks each day
apple_df.loc[apple_df['Open']>= apple_df['Close'], 'Result'] = 0
apple_df.loc[apple_df['Open']< apple_df['Close'], 'Result'] = 1
apple_df
apple_df.to_csv("raw_stock/apple_stock.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np
x_axis = np.arange(len(apple_df))
plt.plot(x_axis,apple_df["Close"])
plt.plot(x_axis,apple_df["Open"])
```
